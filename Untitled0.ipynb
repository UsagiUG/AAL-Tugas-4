```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Generate random numbers
min_data_size = 100_000
num_data_sizes = 30
N = min_data_size*num_data_sizes
arr = np.random.randint(0, N, N)

# Time each sorting algorithm
def time_sort_numpy(num_of_data, kind):
    data = arr[0:num_of_data].copy()
    start = time.time()
    np.sort(data, kind = kind)
    end = time.time()
    return end - start

# Quicksort
arr_of_data_size = np.arange(min_data_size, N+1, min_data_size)
quicksort = []
for i in range(arr_of_data_size.size):
  quicksort.append(time_sort_numpy(arr_of_data_size[i], 'quicksort'))

plt.plot(arr_of_data_size, quicksort, marker="o")
plt.title('Quicksort')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.show()
```

```python
mergesort = []
for i in range(arr_of_data_size.size):
  mergesort.append(time_sort_numpy(arr_of_data_size[i], 'mergesort'))

plt.plot(arr_of_data_size, mergesort, marker="o")
plt.title('Mergesort')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.show()
```

```python
# Generate random numbers
min_data_size = 100   # keep this small for O(n^2) sorts
num_data_sizes = 20
N = min_data_size * num_data_sizes
arr = np.random.randint(0, N, N)

# Insertion sort implementation
def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Timing wrapper
def time_sort_insertion(num_of_data):
    data = arr[0:num_of_data].copy().tolist()
    start = time.time()
    insertion_sort(data)
    end = time.time()
    return end - start

# Input sizes
arr_of_data_size = np.arange(min_data_size, N+1, min_data_size)

# Measure times
insertionsort = []
for i in range(arr_of_data_size.size):
    insertionsort.append(time_sort_insertion(arr_of_data_size[i]))

# Plot
plt.plot(arr_of_data_size, insertionsort, marker="o")
plt.title('Insertion Sort')
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.show()
```

```python
# Generate random numbers
min_data_size = 100   # small so it runs fast
num_data_sizes = 20
N = min_data_size * num_data_sizes
arr = np.random.randint(0, N, N)

def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def time_sort_insertion(num_of_data):
    data = arr[0:num_of_data].copy().tolist()
    start = time.time()
    insertion_sort(data)
    end = time.time()
    return end - start

arr_of_data_size = np.arange(min_data_size, N+1, min_data_size)
insertionsort = [time_sort_insertion(size) for size in arr_of_data_size]

plt.plot(arr_of_data_size, insertionsort, marker="o")
plt.title("Insertion Sort")
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.show()
```

```python
min_data_size = 100
num_data_sizes = 20
N = min_data_size * num_data_sizes
arr = np.random.randint(0, N, N)

def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def time_sort_bubble(num_of_data):
    data = arr[0:num_of_data].copy().tolist()
    start = time.time()
    bubble_sort(data)
    end = time.time()
    return end - start

arr_of_data_size = np.arange(min_data_size, N+1, min_data_size)
bubblesort = [time_sort_bubble(size) for size in arr_of_data_size]

plt.plot(arr_of_data_size, bubblesort, marker="o")
plt.title("Bubble Sort")
plt.xlabel("Input size (n)")
plt.ylabel("Time (seconds)")
plt.show()
```
